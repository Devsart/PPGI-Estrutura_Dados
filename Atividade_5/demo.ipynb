```python
from graph import Graph
```

```python
grafo = {}
kruskal = {}
prim = {}
dijkstra = {}
for i in ['10','20','40','50']:
    grafo[i] = Graph("dij"+i+".txt")
    kruskal[i] = grafo[i].kruskal()
    prim[i] = grafo[i].prim()
    dijkstra[i] = grafo[i].dijkstra('1',i)
    
    print('Resultados para as instâncias do arquivo '+"dij"+i+".txt\n")
    print('Kruskal:\n '+str(kruskal[i])+'\n\n')
    print('Prim:\n '+str(prim[i])+'\n\n')
    print('Dijkstra:\n '+str(dijkstra[i])+'\n\n')
```

```python
def soma_prim_kruskal(i):
    soma_prim=0
    soma_kruskal=0
    for v in prim[i]:
        soma_prim += grafo[i].peso(v)
    for v in kruskal[i]:
        soma_kruskal += grafo[i].peso(v)
    return soma_prim,soma_kruskal
```

```python
soma_prim,soma_kruskal = soma_prim_kruskal('10')
soma_prim,soma_kruskal
```
