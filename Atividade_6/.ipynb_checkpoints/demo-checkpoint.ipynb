```python
from mochila import Mochila
```

```python
import os
diretorio = 'instancias_mochila'
for filename in os.listdir(diretorio):
    f = os.path.join(diretorio, filename)
    if os.path.isfile(f):
        if('soluções' not in f):
            pesos,valores,capacidade,n = Mochila.ler_instancia(f)
            itens, valor = Mochila.getMax(pesos,valores,capacidade)
            _valor = Mochila.getMaxDinamico(pesos,valores,capacidade,n)
            print('Analisando o arquivo - '+f+'\n')
            print('Os itens da mochila inteira são - '+str(itens)+'\n')
            print('O valor total na mochila é - '+str(valor)+'\n')
            print('Algoritmo validado - '+str(valor == _valor)+'.\n\n')
```

### A partir daqui, o código foi alterado para pegar a lista dos itens também a partir do GetMaxDinamico. o Erro do valor máximo no GetMax e GetMaxDinamico na instancia de 2500 foi de apenas 1 unidade, por isso não validou o método, mas o correto é o GetMaxDinâmico, que acusou 37138 como valor de referência.

```python
pesos,valores,capacidade,n = Mochila.ler_instancia('instancias_mochila\mochila01.txt.txt')
```

```python
g,x = Mochila.getMaxDinamico(pesos,valores,capacidade,n)
```

```python
_,l_itens = Mochila.get_items(g,x,pesos)
print(f'Valor máximo - {g.max()}\nLista de Itens na Mochila - {l_itens}')
```
